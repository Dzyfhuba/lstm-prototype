# Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# Load final.csv

```python
df_raw = pd.read_csv('final.csv', index_col='datetime')
df_raw.head()
```

# Perpare columns variable for future usage

```python
columns = ['datetime','choke','psig','apsig','oil','gas','water']
columns2 = ['choke','psig','apsig','oil','gas','water']
```

# Normalization

```python
scaler = MinMaxScaler()
df = scaler.fit_transform(df_raw)
df = pd.DataFrame(df, columns=columns2)
df['datetime'] = df_raw.index
df = df.set_index('datetime')
df.head()
```

# Plot unnormalized and normalized dataframe

```python
df_raw.plot(subplots=True, figsize=(20, 10), title='Unnormalized Dataframe')
df.plot(subplots=True, figsize=(20, 10), title='Normalized Dataframe')
```
